```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("Exercise_Data.csv")

# Show first few rows
print(df.head())

# ----------------------------
# HEATMAP OF PULSE DATA
# ----------------------------
pulse_data = df[["1 min", "15 min", "30 min"]]

plt.figure(figsize=(8, 6))
sns.heatmap(pulse_data.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Heatmap of Pulse Data")
plt.show()

# ----------------------------
# RESHAPE DATA FOR CATEGORICAL PLOTS
# ----------------------------
long_df = df.melt(
    id_vars=["id", "diet", "kind"],
    value_vars=["1 min", "15 min", "30 min"],
    var_name="Time",
    value_name="Pulse"
)

# ----------------------------
# CATEGORICAL PLOT OF PULSE BY DIET
# ----------------------------
sns.catplot(
    data=long_df,
    x="diet",
    y="Pulse",
    hue="Time",
    kind="bar",
    height=6,
    aspect=1.4
)
plt.title("Pulse Values by Diet")
plt.xlabel("Diet Type")
plt.ylabel("Pulse")
plt.show()

# ----------------------------
# CATEGORICAL PLOT OF PULSE BY TYPE OF EXERCISE
# ----------------------------
sns.catplot(
    data=long_df,
    x="kind",
    y="Pulse",
    hue="Time",
    kind="bar",
    height=6,
    aspect=1.6
)
plt.title("Pulse Values by Type of Exercise")
plt.xlabel("Type of Exercise")
plt.ylabel("Pulse")
plt.xticks(rotation=20)
plt.show()

# ----------------------------
# OPTIONAL: BOXPLOT FOR BETTER VISUALS
# ----------------------------
plt.figure(figsize=(10, 6))
sns.boxplot(data=long_df, x="kind", y="Pulse", hue="diet")
plt.title("Pulse Distribution by Exercise Type and Diet")
plt.xlabel("Type of Exercise")
plt.ylabel("Pulse")
plt.xticks(rotation=20)
plt.legend(title="Diet")
plt.show()

# ----------------------------
# CONCLUSION
# ----------------------------
print("\nConclusion:")
print("The data shows that pulse rates change depending on the type of exercise being done.")
print("More active exercise types tend to have higher pulse values than resting.")
print("Diet also seems to affect pulse, but exercise has the stronger effect overall.")
print("This can help show elementary students that physical activity has a clear impact on heart rate.")
```
